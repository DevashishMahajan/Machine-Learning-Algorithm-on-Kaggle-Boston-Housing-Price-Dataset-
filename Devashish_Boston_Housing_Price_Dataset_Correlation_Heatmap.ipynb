```python
from scipy.stats import pearsonr

# Pearson correlation coefficient between "lstat" and "medv"
pearsonr(boston['lstat'],boston['medv'])
```

```python
# Pearson correlation coefficient between "ptratio" and "medv"
pearsonr(boston['ptratio'],boston['medv'])
```

```python
# Pearson correlation coefficient between "rm" and "ptration"
pearsonr(boston['rm'],boston['ptratio'])
```

```python
# heatmap of Pearson correlation coefficient
plt.figure (figsize= (12,12))
sns.heatmap(
    boston.corr(),
    
    annot=True)
plt.show()
```
